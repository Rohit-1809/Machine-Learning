```python
#PHASE 1 :Data Preprocessing and Feature Engineering
# data import and cleaning 
# handle missing values through through imputation 

import pandas as pd

df = pd.read_csv("Food_Delivery_Time_Prediction.csv")

df_info = df.info()
missing_values = df.isnull().sum()

df.head(), missing_values

```

```python
# Encoding categorial variables .LabelEncoder
# normalization of numerical feature .MinMaxscaler.

from sklearn.preprocessing import OneHotEncoder, LabelEncoder, MinMaxScaler

df_encoded = df.copy()

categorical_columns_onehot = ['Weather_Conditions', 'Traffic_Conditions', 'Order_Priority', 'Order_Time', 'Vehicle_Type']
label_encoder_columns = []  # All chosen columns will use One-Hot Encoding

df_encoded = pd.get_dummies(df_encoded, columns=categorical_columns_onehot)

numerical_columns = ['Distance', 'Delivery_Person_Experience', 'Restaurant_Rating', 
                     'Customer_Rating', 'Delivery_Time', 'Order_Cost', 'Tip_Amount']


scaler = MinMaxScaler()
df_encoded[numerical_columns] = scaler.fit_transform(df_encoded[numerical_columns])


df_encoded.head()
```

```python
# feature engineering :
# calculation of geographical distance . haversine  formula 
# deriving time related features "rush hour" or "non-rush hour".

import numpy as np

def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371  
    phi1 = np.radians(lat1)
    phi2 = np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)

    a = np.sin(delta_phi / 2.0) ** 2 + \
        np.cos(phi1) * np.cos(phi2) * np.sin(delta_lambda / 2.0) ** 2

    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    return R * c


def extract_coordinates(coord_str):
    coord_str = coord_str.strip('()')
    lat_str, lon_str = coord_str.split(',')
    return float(lat_str), float(lon_str)


customer_coords = df['Customer_Location'].apply(extract_coordinates)
restaurant_coords = df['Restaurant_Location'].apply(extract_coordinates)


df['Customer_Lat'], df['Customer_Lon'] = zip(*customer_coords)
df['Restaurant_Lat'], df['Restaurant_Lon'] = zip(*restaurant_coords)


df['Haversine_Distance'] = haversine_distance(
    df['Customer_Lat'], df['Customer_Lon'],
    df['Restaurant_Lat'], df['Restaurant_Lon']
)


def is_rush_hour(order_time):
    return 1 if order_time in ['Morning', 'Evening'] else 0

df['Rush_Hour'] = df['Order_Time'].apply(is_rush_hour)


df[['Distance', 'Haversine_Distance', 'Order_Time', 'Rush_Hour']].head()
```

```python
# PHASE 2 : Clustering using K-Means and Hierarchical Clustering.

#  1 )K-Means Clustering.

# objectives and implementation : k-means algorithm and elbow method to determine optimal number of clusters .
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns



df_encoded['Haversine_Distance'] = df['Haversine_Distance']
df_encoded['Rush_Hour'] = df['Rush_Hour']


from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_encoded[['Haversine_Distance', 'Rush_Hour']] = scaler.fit_transform(
    df_encoded[['Haversine_Distance', 'Rush_Hour']]
)


features_for_clustering = df_encoded.drop(columns=['Order_ID', 'Customer_Location', 'Restaurant_Location'])


inertia = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features_for_clustering)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method - Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

# KMeans we choose k=3 here  
kmeans = KMeans(n_clusters=3, random_state=42)
df_encoded['Cluster'] = kmeans.fit_predict(features_for_clustering)


pca = PCA(n_components=2)
pca_result = pca.fit_transform(features_for_clustering)
df_encoded['PCA1'] = pca_result[:, 0]
df_encoded['PCA2'] = pca_result[:, 1]

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_encoded, x='PCA1', y='PCA2', hue='Cluster', palette='Set2')
plt.title('K-Means Clustering Visualization (PCA)')
plt.show()
```

```python
# Evaluaion :
#Visualize clusters using 2D or 3D scatter plots.

from mpl_toolkits.mplot3d import Axes3D

pca_3d = PCA(n_components=3)
pca_components = pca_3d.fit_transform(features_for_clustering)

df_encoded['PCA3'] = pca_components[:, 2]


fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(
    df_encoded['PCA1'],
    df_encoded['PCA2'],
    df_encoded['PCA3'],
    c=df_encoded['Cluster'],
    cmap='Set1',
    alpha=0.8
)
ax.set_title('3D PCA - KMeans Clustering')
ax.set_xlabel('PCA1')
ax.set_ylabel('PCA2')
ax.set_zlabel('PCA3')
plt.colorbar(scatter, label='Cluster')
plt.show()

```

```python
#Analyze how delivery times are influenced by different clusters.


cluster_analysis = df_encoded.groupby('Cluster')['Delivery_Time'].agg(['mean', 'median', 'count']).reset_index()
cluster_analysis.rename(columns={'mean': 'Avg_Delivery_Time', 'median': 'Median_Delivery_Time'}, inplace=True)
print(cluster_analysis)


plt.figure(figsize=(8, 6))
sns.boxplot(data=df_encoded, x='Cluster', y='Delivery_Time', palette='Set2')
plt.title('Delivery Time Distribution by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Normalized Delivery Time')
plt.grid(True)
plt.show()
```

```python
# 2) Hierarchical Clustering : 
# objectives and implementation : agglomerative clustering and we use dendrograms to visualize clusters .

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
import numpy as np


def extract_coordinates(location_str):
    lat, lon = location_str.strip("()").split(", ")
    return float(lat), float(lon)


df[['Customer_Lat', 'Customer_Lon']] = df['Customer_Location'].apply(lambda x: pd.Series(extract_coordinates(x)))
df[['Restaurant_Lat', 'Restaurant_Lon']] = df['Restaurant_Location'].apply(lambda x: pd.Series(extract_coordinates(x)))


features = df[[
    'Distance', 'Delivery_Person_Experience', 'Restaurant_Rating',
    'Customer_Rating', 'Order_Cost', 'Tip_Amount',
    'Customer_Lat', 'Customer_Lon', 'Restaurant_Lat', 'Restaurant_Lon',
    'Weather_Conditions', 'Traffic_Conditions', 'Order_Priority',
    'Order_Time', 'Vehicle_Type'
]]


numerical_features = features.select_dtypes(include=['float64', 'int64']).columns.tolist()
categorical_features = features.select_dtypes(include=['object']).columns.tolist()


preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_features),
    ('cat', OneHotEncoder(drop='first'), categorical_features)
])


X_preprocessed = preprocessor.fit_transform(features)


linked = linkage(X_preprocessed, method='ward')


plt.figure(figsize=(12, 6))
dendrogram(linked, truncate_mode='level', p=5)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index or (Cluster Size)')
plt.ylabel('Distance')
plt.tight_layout()
plt.show()

```

```python
# Compare cluster analysis with K-Means results to find patterns.


from sklearn.cluster import AgglomerativeClustering, KMeans
from sklearn.decomposition import PCA
import seaborn as sns


agg_cluster = AgglomerativeClustering(n_clusters=4)
agg_labels = agg_cluster.fit_predict(X_preprocessed)


kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(X_preprocessed)


pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_preprocessed)


cluster_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
cluster_df['Agglomerative'] = agg_labels
cluster_df['KMeans'] = kmeans_labels


plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.scatterplot(data=cluster_df, x='PC1', y='PC2', hue='Agglomerative', palette='Set2')
plt.title('Agglomerative Clustering (4 Clusters)')
plt.legend(title='Cluster')


plt.subplot(1, 2, 2)
sns.scatterplot(data=cluster_df, x='PC1', y='PC2', hue='KMeans', palette='Set1')
plt.title('K-Means Clustering (4 Clusters)')
plt.legend(title='Cluster')

plt.tight_layout()
plt.show()
```

```python
# Left: Agglomerative Clustering (Hierarchical) we took 4 clusters .
# Right: K-Means Clustering

#Observations: comparison of cluster analysis 
#1) Both methods identify similar grouping patterns in the 2D PCA-reduced space, but the cluster boundaries differ slightly.

#2)Agglomerative Clustering may find more naturally shaped clusters due to its hierarchical nature.

#3)K-Means Clustering assumes spherical clusters and tries to minimize variance within each.
```

```python
# PHASE 3 : Neural network for prediction .


# objectives : predict if delivery will be fast or delayed using neural network 

#Fast (0) if Delivery_Time ≤ median
#Delayed (1) if Delivery_Time > median
#we already preprocessed the data: categorical variables are encoded, numerical features are scaled.
#the target is binary (0 for "Fast", 1 for "Delayed").

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import numpy as np


median_delivery_time = df['Delivery_Time'].median()
df['Delivery_Status'] = (df['Delivery_Time'] >= median_delivery_time).astype(int)  # 0 = Fast, 1 = Delayed

df_cleaned = df.drop(columns=[
    'Order_ID',
    'Customer_Location',
    'Restaurant_Location',
    'Delivery_Time' 
])


X = df_cleaned.drop(columns=['Delivery_Status'])
y = df_cleaned['Delivery_Status']


categorical_cols = ['Weather_Conditions', 'Traffic_Conditions', 'Order_Priority', 'Order_Time', 'Vehicle_Type']
numerical_cols = [col for col in X.columns if col not in categorical_cols]


preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
])


X_processed = preprocessor.fit_transform(X)


X_train, X_test, y_train, y_test = train_test_split(
    X_processed, y, test_size=0.2, random_state=42
)

X_processed.shape, X_train.shape, X_test.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense


model = Sequential([
    Dense(32, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')  # Sigmoid for binary classification
])


model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

#
history = model.fit(X_train, y_train, epochs=50, batch_size=16, validation_split=0.2, verbose=0)


test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
test_loss, test_accuracy
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense


df = pd.read_csv("Food_Delivery_Time_Prediction.csv")


median_time = df['Delivery_Time'].median()
df['Delivery_Status'] = (df['Delivery_Time'] >= median_time).astype(int)

df = df.drop(columns=['Order_ID', 'Customer_Location', 'Restaurant_Location', 'Delivery_Time'])


X = df.drop(columns=['Delivery_Status'])
y = df['Delivery_Status']


cat_cols = ['Weather_Conditions', 'Traffic_Conditions', 'Order_Priority', 'Order_Time', 'Vehicle_Type']
num_cols = [col for col in X.columns if col not in cat_cols]


preprocessor = ColumnTransformer([
    ('num', StandardScaler(), num_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
])
X_processed = preprocessor.fit_transform(X)


X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42)


model = Sequential([
    Dense(32, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])


model.fit(X_train, y_train, epochs=50, batch_size=16, validation_split=0.2)


loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy:.2f}")
```

```python
# Evaluation matrics : 

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob >= 0.5).astype(int)


accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy : {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall   : {recall:.2f}")
print(f"F1 Score : {f1:.2f}")
```

```python
# model improvement :
# using tuned hyperparameters 
from tensorflow.keras.optimizers import Adam

# Build a tuned neural network
model = Sequential([
    Dense(64, activation='tanh', input_shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile with custom learning rate
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])


model.fit(X_train, y_train, epochs=100, batch_size=16, validation_split=0.2)
```

```python
#Evaluate the neural network's performance compared to traditional machine learning models (e.g., Logistic Regression).


from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam


lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train, y_train)
y_lr_pred = lr_model.predict(X_test)


nn_model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
nn_model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
nn_model.fit(X_train, y_train, epochs=50, batch_size=16, validation_split=0.2, verbose=0)

y_nn_pred_prob = nn_model.predict(X_test)
y_nn_pred = (y_nn_pred_prob >= 0.5).astype(int)


def evaluate(name, y_true, y_pred):
    print(f"\n{name} Performance:")
    print(f"Accuracy : {accuracy_score(y_true, y_pred):.2f}")
    print(f"Precision: {precision_score(y_true, y_pred):.2f}")
    print(f"Recall   : {recall_score(y_true, y_pred):.2f}")
    print(f"F1 Score : {f1_score(y_true, y_pred):.2f}")

# 4. Comparison
evaluate("Logistic Regression", y_test, y_lr_pred)
evaluate("Neural Network", y_test, y_nn_pred)
```

```python
# PHASE 4) Reporting and insights :
# model comparison : comparing custering result and neural network accuracy . 
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np


kmeans = KMeans(n_clusters=2, random_state=42)
cluster_labels = kmeans.fit_predict(X_test)


accuracy_option1 = accuracy_score(y_test, cluster_labels)
accuracy_option2 = accuracy_score(y_test, 1 - cluster_labels)


if accuracy_option2 > accuracy_option1:
    cluster_labels = 1 - cluster_labels


def evaluate(name, y_true, y_pred):
    print(f"\n{name} Performance:")
    print(f"Accuracy : {accuracy_score(y_true, y_pred):.2f}")
    print(f"Precision: {precision_score(y_true, y_pred):.2f}")
    print(f"Recall   : {recall_score(y_true, y_pred):.2f}")
    print(f"F1 Score : {f1_score(y_true, y_pred):.2f}")


evaluate("Neural Network", y_test, y_nn_pred)
evaluate("KMeans Clustering", y_test, cluster_labels)
```
